```python
## store 1
NofDate=366
NofReq=100
productID=range(1000,2000)
NofPrdMax=50
fpath_order = os.path.join('dataH', '2023order.csv')


date_list=[datetime.datetime(2023, 4, 1)+datetime.timedelta(days=i) for i in range(NofDate)]
display(date_list[0])
display(date_list[-1])

NofDreq=np.random.randint(1,NofReq, (NofDate,1))

for k in range(NofDate):
    r=set()
    while len(r) < NofDreq[k]:
        k2=len(r)
        n = np.random.randint(1000,2000)
        r.add(n)
    dP=np.vstack([np.full(NofDreq[k],k),np.array(list(r)),np.random.randint(1,NofPrdMax,NofDreq[k])])
#     display(dP)
#     display(dP.T)
    if k==0:
        data_full=dP.T
    else:
         data_full=np.vstack([data_full,dP.T])
    
df_order=pd.DataFrame(data_full,columns=['date', 'productID', 'order'])
df_order['date']=df_order['date'].map(lambda x: date_list[x])
df_order.to_csv(fpath_order)
```
